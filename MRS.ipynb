```python
def convert(text):
    L = []
    try:
        # Safely evaluate the string to a Python object
        data = ast.literal_eval(text)
        # Ensure data is a list
        if isinstance(data, list):
            for item in data:
                # Check if each item is a dictionary with a 'name' key
                if isinstance(item, dict) and 'name' in item:
                    L.append(item['name'])
             
        else:
            print(f"Not a list: {data}")
    except (ValueError, SyntaxError) as e:
        print(f"Error processing text: {text}. Error: {e}")
    return L
```

```python
movies['cast'] = movies['cast'].apply(lambda x:x[0:3])
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
# vectorization
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
movies_vector = cv.fit_transform(new_df['tags']).toarray()
movies_vector[0]
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(movies_vector)
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

def stem(text):
    y = []
    for i in text.split():
      y.append(ps.stem(text))

    return " ".join(y)
```

```python
def recommend(movie_title):
    filtered_df = new_df[new_df['title'].str.lower() == movie_title.lower()]
    if not filtered_df.empty:
        index = filtered_df.index[0]
        distances = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda x: x[1])
        for i in distances[1:6]:
            print(new_df.iloc[i[0]].title)
    else:
        print(f"No matches found for '{movie_title}'.")
```
